```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("International_T20_Data.csv")
df.head()
```

```python
##BASIC OPERATIONS OF PANDAS
df.head()
df.tail()
df.sample(5)
df.shape
df.columns
df.index
df.dtypes
df.info()
df.describe()
```

```python
#DATA CLEANING
##-MISSING VALUES
df.isnull().sum()
# df.dropna(inplace=True) # Commenting out to prevent loss of all data

##Removing duplicates
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
##COLUMN OPERATIONS
df.rename(columns={'info.city':'City'}, inplace=True)

df['Runs_margin'] = df['info.outcome.by.runs']

df.sort_values(by='Runs_margin', ascending=False).head()
```

```python

# Preprocess Column Names
import ast

# Rename columns
def rename_columns(df):
    new_columns = []
    for col in df.columns:
        col = col.replace('.', '_')
        if col.startswith('info_'):
            col = col[len('info_'):]
        elif col.startswith('meta_'):
            col = col[len('meta_'):]
        new_columns.append(col)
    df.columns = new_columns
    return df

df = rename_columns(df)

# Convert 'innings' column
df['innings'] = df['innings'].apply(ast.literal_eval)

df.head()
df.info()
```

```python
# Top Venues
top_venues = df['venue'].value_counts().head(3)
print("Top Three Venues by Number of Matches:")
print(top_venues)
```

```python
# Most Played Team Pair
import ast

df['teams'] = df['teams'].apply(ast.literal_eval)

df['sorted_teams'] = df['teams'].apply(lambda x: tuple(sorted(x)))

team_pair_counts = df['sorted_teams'].value_counts()

most_played_pair = team_pair_counts.index[0]
num_matches = team_pair_counts.iloc[0]

print(f"The team pair that has played the most T20 matches against each other is {most_played_pair} with {num_matches} matches.")
```

```python
# Team Win Percentages
all_teams = []
for teams_list in df['teams']:
    all_teams.extend(teams_list)
unique_teams = sorted(list(set(all_teams)))

total_matches = {team: 0 for team in unique_teams}
matches_won = {team: 0 for team in unique_teams}

for index, row in df.iterrows():
    teams_in_match = row['teams']
    winner = row['outcome_winner']

    for team in teams_in_match:
        total_matches[team] += 1

    if pd.notna(winner) and winner in matches_won:
        matches_won[winner] += 1

# Calculate win percentages
win_percentages = {}
for team in unique_teams:
    if total_matches[team] > 0:
        win_percentages[team] = (matches_won[team] / total_matches[team]) * 100
    else:
        win_percentages[team] = 0.0

# Create a pandas Series from the win percentages and sort
win_percentages_series = pd.Series(win_percentages)
top_five_teams = win_percentages_series.sort_values(ascending=False).head(5)

print("Top Five Teams by Win Percentage:")
print(top_five_teams)
```

```python
# Scorecard Function
def generate_scorecard(innings_data):
    scorecards = {}

    for innings_key, innings_details in innings_data[0].items():
        team_name = innings_details['team']

        # Extract batting data
        batting_stats = []
        for delivery in innings_details['deliveries']:
            for ball_data in delivery.values():
                batsman = ball_data['batsman']
                runs = ball_data['runs']['batsman']
                batting_stats.append({'player': batsman, 'runs': runs})

        batsman_runs = pd.DataFrame(batting_stats).groupby('player')['runs'].sum().reset_index()
        top_batsmen = batsman_runs.sort_values(by='runs', ascending=False).head(4)

        # Extract bowling data
        bowling_stats = []
        for delivery in innings_details['deliveries']:
            for ball_data in delivery.values():
                bowler = ball_data['bowler']
                wickets = 1 if 'wicket' in ball_data else 0
                runs_conceded = ball_data['runs']['total'] # Total runs conceded on that ball
                bowling_stats.append({'player': bowler, 'wickets': wickets, 'runs_conceded': runs_conceded})

        bowler_df = pd.DataFrame(bowling_stats)
        bowler_wickets = bowler_df.groupby('player')['wickets'].sum().reset_index()
        bowler_runs = bowler_df.groupby('player')['runs_conceded'].sum().reset_index()

        bowler_summary = pd.merge(bowler_wickets, bowler_runs, on='player', how='left')
        top_bowlers = bowler_summary.sort_values(by=['wickets', 'runs_conceded'], ascending=[False, True]).head(4)

        # Create DataFrame for current innings scorecard
        scorecard_df = pd.DataFrame({
            'Batsman': top_batsmen['player'].tolist(),
            'Runs': top_batsmen['runs'].tolist(),
            'Bowler': top_bowlers['player'].tolist(),
            'Wickets': top_bowlers['wickets'].tolist()
        })
        scorecards[innings_key] = scorecard_df

    # Returning two dataframes, one for '1st innings' and one for '2nd innings'
    innings_1_key = list(innings_data[0].keys())[0]
    innings_2_key = list(innings_data[0].keys())[1]

    return scorecards[innings_1_key], scorecards[innings_2_key]

print("generate_scorecard function defined.")
```

```python
def generate_scorecard(innings_list_of_dicts):
    scorecards = {}

    for innings_entry in innings_list_of_dicts:
        innings_key = list(innings_entry.keys())[0] # Get '1st innings' or '2nd innings'
        innings_details = innings_entry[innings_key]

        team_name = innings_details['team']

        # Extract batting data
        batting_stats = []
        for delivery in innings_details['deliveries']:
            for ball_data in delivery.values():
                batsman = ball_data['batsman']
                runs = ball_data['runs']['batsman']
                batting_stats.append({'player': batsman, 'runs': runs})

        batsman_runs = pd.DataFrame(batting_stats).groupby('player')['runs'].sum().reset_index()
        top_batsmen = batsman_runs.sort_values(by='runs', ascending=False).head(4)

        # Extract bowling data
        bowling_stats = []
        for delivery in innings_details['deliveries']:
            for ball_data in delivery.values():
                bowler = ball_data['bowler']
                wickets = 1 if 'wicket' in ball_data else 0
                runs_conceded = ball_data['runs']['total'] # Total runs conceded on that ball
                bowling_stats.append({'player': bowler, 'wickets': wickets, 'runs_conceded': runs_conceded})

        bowler_df = pd.DataFrame(bowling_stats)
        bowler_wickets = bowler_df.groupby('player')['wickets'].sum().reset_index()
        bowler_runs = bowler_df.groupby('player')['runs_conceded'].sum().reset_index()

        bowler_summary = pd.merge(bowler_wickets, bowler_runs, on='player', how='left')
        top_bowlers = bowler_summary.sort_values(by=['wickets', 'runs_conceded'], ascending=[False, True]).head(4)

        # DataFrame for current innings scorecard
        scorecard_df = pd.DataFrame({
            'Batsman': top_batsmen['player'].tolist(),
            'Runs': top_batsmen['runs'].tolist(),
            'Bowler': top_bowlers['player'].tolist(),
            'Wickets': top_bowlers['wickets'].tolist()
        })
        scorecards[innings_key] = scorecard_df

    # Returning two dataframes, one for '1st innings' and one for '2nd innings'
    innings_1_key = list(innings_list_of_dicts[0].keys())[0]
    innings_2_key = list(innings_list_of_dicts[1].keys())[0]

    return scorecards[innings_1_key], scorecards[innings_2_key]

print("generate_scorecard function defined.")
```
